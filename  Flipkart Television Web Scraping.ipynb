```python
import requests
from bs4 import BeautifulSoup
import csv

data = requests.get("https://www.flipkart.com/search?q=televisions")
# print(data.content)
soup = BeautifulSoup(data.content, "html.parser")
# print(soup.prettify())
# print(soup.title)
img = print(soup.find_all)
print(img)
```

```python
products = soup.find_all('div',class_= "_3pLy-c")
for single in products:
    title = single.find("div",class_="_4rR01T")
    print(title.text)
#     print(title)
```

```python
products = soup.find_all('div',class_= "_3pLy-c")
for single in products:
    title = single.find("div",class_="_4rR01T")
    price = single.find("div", class_="_30jeq3")
    print()
    print(title.text,"\t",price.text)
```

```python
import csv
file = open("products.csv","w",newline='',encoding="utf-8")
wr = csv.writer(file)
wr.writerow(["product name","Product Price"])
products = soup.find_all('div',class_= "_3pLy-c")
for single in products:
    title = single.find("div",class_="_4rR01T")
    price = single.find("div", class_="_30jeq3")
    wr.writerow([title.text,price.text])
    
file.close()
```
